```python
import cv2
import matplotlib.pyplot as plt

# Function to apply image smoothing and edge detection
def enhance_and_detect_edges(image_path):
    # Read the noisy image
    noisy_image = cv2.imread(image_path)

    # Apply image smoothing (GaussianBlur) for denoising
    denoised_image = cv2.GaussianBlur(noisy_image, (5, 5), 0)  # You can adjust the kernel size (e.g., (5, 5), (10, 10), etc.)

    # Apply edge detection (Canny edge detector)
    edges = cv2.Canny(denoised_image, 100, 200)  # You can adjust the threshold values (e.g., 100, 200) to get desired edges.

    return denoised_image, edges

# File paths for the noisy images
fighter_noise_5_path = "/Users/benedictadadson/Downloads/fighter_noise_5.jpg"
fighter_noise_10_path = "/Users/benedictadadson/Downloads/fighter_noise_10.jpg"

# Apply image smoothing and edge detection for the first image
denoised_image_5, edges_5 = enhance_and_detect_edges(fighter_noise_5_path)

# Apply image smoothing and edge detection for the second image
denoised_image_10, edges_10 = enhance_and_detect_edges(fighter_noise_10_path)

# Display the results
plt.figure(figsize=(12, 6))

plt.subplot(2, 4, 1)
plt.imshow(cv2.cvtColor(cv2.imread(fighter_noise_5_path), cv2.COLOR_BGR2RGB))
plt.title("Noisy Image 5")

plt.subplot(2, 4, 2)
plt.imshow(cv2.cvtColor(denoised_image_5, cv2.COLOR_BGR2RGB))
plt.title("Denoised Image 5")

plt.subplot(2, 4, 3)
plt.imshow(edges_5, cmap="gray")
plt.title("Edge Image 5")

plt.subplot(2, 4, 5)
plt.imshow(cv2.cvtColor(cv2.imread(fighter_noise_10_path), cv2.COLOR_BGR2RGB))
plt.title("Noisy Image 10")

plt.subplot(2, 4, 6)
plt.imshow(cv2.cvtColor(denoised_image_10, cv2.COLOR_BGR2RGB))
plt.title("Denoised Image 10")

plt.subplot(2, 4, 7)
plt.imshow(edges_10, cmap="gray")
plt.title("Edge Image 10")

plt.show()
```
